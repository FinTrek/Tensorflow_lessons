```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
tf.reset_default_graph() ##resetting
```

```python
input_data = tf.placeholder(dtype = tf.float32, shape = None) ##creating an input placeholder
output_data = tf.placeholder(dtype = tf.float32, shape = None) ##creating an output placeholder
a = tf.Variable(1, dtype = tf.float32) ##creates a in parabola
b = tf.Variable(1, dtype = tf.float32) 
c = tf.Variable(1, dtype = tf.float32) 
```

```python
model_operation = a * (input_data ** 2) + (b * input_data) + c ##parabola equation
error = model_operation - output_data ##calculates error
squared_error = tf.square(error)##squares error
##creates and squares errors

loss = tf.reduce_mean(squared_error) ##gets mean of squared error

learning_rate = 0.005
##required stuff that don't 100 percent understand
optimizer = tf.train.GradientDescentOptimizer(learning_rate)
train = optimizer.minimize(loss)
init = tf.global_variables_initializer()
##required stuff that don't 100 percent understand

x_values = np.arange(-5, 6)
print(x_values)                     
y_values = np.square(x_values)

with tf.Session() as sess: 
    sess.run(init)  ##required
    for i in range (2000):  
        sess.run(train, feed_dict = {input_data:x_values, output_data:y_values})  ##putting values into place holder
        if i % 10 == 0:  ##feeding
            print(sess.run([a, b, c]))
            plt.plot(x_values, sess.run(model_operation, feed_dict={input_data:x_values})) ##plots point
        
    print(sess.run(loss, feed_dict={input_data:x_values, output_data:y_values}))  
    plt.plot(x_values, y_values, 'ro','Training Data')
    plt.plot(x_values, sess.run(model_operation, feed_dict={input_data:x_values}))
        
    plt.show()
```
