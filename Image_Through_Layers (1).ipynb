```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2

gf = cv2.imread("C:\\Users\\silve\\OneDrive\\Desktop\\flavortown.jpg")
#print(china)

gf = cv2.cvtColor(gf, cv2.COLOR_BGR2RGB)
plt.imshow(gf)
data = np.array([gf], dtype = np.float32)
#print(china)
height, width, channels = gf.shape
#plt.show()

#plt.imshow(data[0,:,:,0])
plt.show()
```

```python
with tf.device('/cpu:0'):
    X = tf.placeholder(tf.float32, shape=(None, height, width, channels))
    convolution = tf.layers.conv2d(X, filters = 32, kernel_size = [5, 5], activation = tf.nn.relu )
    pool_layer = tf.layers.max_pooling2d(convolution, pool_size = [2, 2], strides = [2, 2])
```

```python
init = tf.global_variables_initializer()
with tf.Session(config=tf.ConfigProto(allow_soft_placement=True, log_device_placement=True)) as sess:
    sess.run(init)
    output =sess.run(convolution, feed_dict={X:data})
    output1 = sess.run(pool_layer, feed_dict={X:data})
plt.imshow(output[0, :, :, 1])
plt.show()
plt.imshow(output1[0, :, :, 1])
plt.show()
```
